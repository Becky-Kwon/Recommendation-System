```python
tripadvisor2 = pd.read_excel('tripadvisor-user-data-with-place.xlsx')
id_set2 = set(list(tripadvisor2['place_id']))
id_set2 = sorted(id_set2)
tripadvisor2
```

```python
tripadvisor = pd.read_excel('tripadvisor-user-data-with-place_none삭제.xlsx')
id_set = set(list(tripadvisor['place_id']))
id_set = sorted(id_set)
tripadvisor
```

```python
# none을 지우지않은 review 데이터 셋
len(id_set2)
```

```python
placc_name = []
review = []
rating = []
count = []
id_num = []
```

```python
for KK in id_set2:
    place_review = tripadvisor2[tripadvisor2['place_id']==KK]
    
    review_sum = []
    avg_rating = 0

    for i in range (len(place_review)):

        # 리뷰 합치기
        review_sum.append(place_review.iloc[i,3])

        # 평균 평점은?
        avg_rating += place_review.iloc[i,2]/len(place_review)

        # 총 리뷰수는?
        count_review = len(place_review)
    print(KK,'\n')
    print('모든 리뷰를 모으면\n',review_sum)
    print('평균 평점은' ,avg_rating)
    print(count_review, '개의 리뷰 모으기 완료\n')
    print('-----------------'*7)
    
    id_num.append(KK)
    placc_name.append(place_lilist[place_lilist['id']==KK]['place'].values.astype(np.object))
    review.append(review_sum)
    rating.append(avg_rating)
    count.append(count_review)
```

```python
total_table = pd.DataFrame({'id': id_num, 'place': placc_name, 'count_review':count,'avg_review': rating, 'total_review':review })
for a in range (len(total_table)):
    total_table.iloc[a,1] = str(total_table.iloc[a,1])[2:-2]
total_table
```
